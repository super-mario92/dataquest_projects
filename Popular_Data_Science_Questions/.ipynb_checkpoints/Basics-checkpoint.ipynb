## Popular Data Science Questions

I investigate the Data Science website of Stack Exchange (https://datascience.stackexchange.com) in order to determine the "best" Data Science content (for a website, blog, etc) to write about. 

Their content is stored in a databased which can be accessed via Transact-SQL queries here:
- https://data.stackexchange.com/datascience/query/new

Promising tables:
- Posts (contains scores, views and tags of user posts)
- Users (contains reputation of users)
- Tags (detailed describtion and meta information about tags)



For this analysis I am only interested in the posts from 2019 which are of PostTypeId = 1, which represents a question.

### Read in the extracted file into a dataframe and explore the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
questions_2019 = pd.read_csv("2019_questions.csv")
questions_2019
```

The data is complete except for the FavoriteCount column. Here we have 7432 missing values. We could fix it by imputing zero values.

In the tag column the "><" chars should be replaced by ",".

The Creation Date column should rather be in datetime format.

The FavoriteCount should rather be an int instead of float.

### Cleaning the Data

```python
questions_2019.FavoriteCount.fillna(0, inplace = True)
questions_2019['FavoriteCount'] = questions_2019['FavoriteCount'].astype(int)
questions_2019['CreationDate'] = pd.to_datetime(questions_2019['CreationDate'])
questions_2019.info()
```

```python
questions_2019['Tags'] = questions_2019['Tags'].str.replace("^<|>$", "").str.split("><")
questions_2019['Tags'].sample(10)
```

### Most Used and Most Viewed

```python
tag_dict = {}
# iterate through all column elements, which returns a list for each row
for row_element in questions_2019['Tags']:
    for single_element in row_element:
        if single_element in tag_dict:
            tag_dict[single_element] += 1
        else:
            tag_dict[single_element] = 1    
```

```python
used_tag_df = pd.DataFrame.from_dict(tag_dict, orient="index")
used_tag_df.rename(columns={0: "Used_Count"}, inplace=True)
used_tag_df.sort_values(by = 'Used_Count', inplace = True, ascending = False)
used_tag_df.head(10)
```

```python
view_dict = {}
for row_number in range(0,len(questions_2019['Tags'])):
    for single_element in questions_2019['Tags'][row_number]:
        if single_element in view_dict:
            view_dict[single_element] = view_dict[single_element] + questions_2019['ViewCount'][row_number]
        else:
            view_dict[single_element] = questions_2019['ViewCount'][row_number]
```

```python
view_tag_df = pd.DataFrame.from_dict(view_dict, orient="index")
view_tag_df.rename(columns={0: "View_Count"}, inplace=True)
view_tag_df.sort_values(by = 'View_Count', inplace = True, ascending = False)
view_tag_df.head(10)
```

```python
used_tag_df.head(10).plot(kind="barh", figsize=(16,8))
```

```python
view_tag_df.head(10).plot(kind="barh", figsize=(16,8))
```

### Relations Between Tags

Keras, scikit-learn, TensorFlow are all Python libraries that allow their users to employ deep learning (a type of neural network).

Most of the top tags are all intimately related with one central machine learning theme: deep learning. I can suggest the creation of Python content that uses deep learning for classification problems (and other variations of this suggestion).

### Just a Fad? (Analyze interest in Deep Learning over time)

```python
all_questions = pd.read_csv('all_questions.csv')
```

```python
all_questions['CreationDate'] = pd.to_datetime(all_questions['CreationDate'])
```

```python
all_questions['Tags'] = all_questions['Tags'].str.replace("^<|>$", "").str.split("><")
all_questions['Tags'].sample(10)
```

```python
# add a column that defines if a questions is about deep learning
def class_deep_learning(tags):
    for tag in tags:
        if tag in ["lstm", "cnn", "scikit-learn", "tensorflow",
                   "keras", "neural-network", "deep-learning"]:
            return 1
    return 0
```

```python
all_questions["DeepLearning"] = all_questions["Tags"].apply(class_deep_learning)
```

```python
all_questions.sample(10)
```

```python
# drop data for year 2020
all_questions = all_questions[all_questions["CreationDate"].dt.year < 2020]
```

```python
# identify quarter in which a question was asked
def fetch_quarter(datetime):
    year = str(datetime.year)[-2:]
    quarter = str(((datetime.month-1) // 3) + 1)
    return "{y}Q{q}".format(y=year, q=quarter)

all_questions["Quarter"] = all_questions["CreationDate"].apply(fetch_quarter)
```

```python
all_questions.sample(10)
```

```python
quarterly = all_questions.groupby("Quarter").agg({"DeepLearning": ["sum", "size"]})
print(quarterly.head())
quarterly.columns = ["DeepLearningQuestions", "TotalQuestions"]
quarterly["DeepLearningRate"] = quarterly["DeepLearningQuestions"] / quarterly["TotalQuestions"]
# The following is done to help with visualizations later.
quarterly.reset_index(inplace=True)
quarterly.sample(5)
```

```python
ax1 = quarterly.plot(x="Quarter", y="DeepLearningRate",
                    kind="line", linestyle="-", marker="o", color="orange",
                    figsize=(30,16)
                    )

ax2 = quarterly.plot(x="Quarter", y="TotalQuestions",
                     kind="bar", ax=ax1, secondary_y=True, alpha=0.7, rot=45)

for idx, t in enumerate(quarterly["TotalQuestions"]):
    ax2.text(idx, t, str(t), ha="center", va="bottom")
xlims = ax1.get_xlim()

ax1.get_legend().remove()

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles=handles1 + handles2,
           labels=labels1 + labels2,
           loc="upper left", prop={"size": 12})


for ax in (ax1, ax2):
    for where in ("top", "right"):
        ax.spines[where].set_visible(False)
        ax.tick_params(right=False, labelright=False)
```

Seems as if deep learning interest increased tremendously and is now rather stagnating but there is no indication that interest is descreasing. 

Thus I stick to my recommendation on choosing deep learning as content topic.
